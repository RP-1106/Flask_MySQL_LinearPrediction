```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('insurance.csv')
data.head(5)
```

```python
data['region'] = data['region'].astype('category')
data['region'] = data['region'].cat.codes
data.head(5)
```

```python
df2 = pd.get_dummies(data['smoker'])
df2 = pd.DataFrame(df2)
data = pd.concat([data,df2],axis=1)
data.head(5)
```

```python
data = data.drop(['smoker','no'],axis=1)
data.rename(columns={'yes':'smoker'},inplace=True)
data.head(5)
```

```python
df2 = pd.get_dummies(data['sex'])
df2 = pd.DataFrame(df2)
data = pd.concat([data,df2],axis=1)
data.head(5)
```

```python
data = data.drop(['sex','male'],axis=1)
data.rename(columns={'female':'sex'},inplace=True)
data.head(5)
```

```python
col=["age","sex","bmi","children","smoker","region","charges"]
data = data[col]
data.head(5)
```

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x,y)
```

```python
result=regression.predict( [ [19,1,27.9,0,1,3] ] )
```
